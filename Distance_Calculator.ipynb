# Distance Calculator


This application aims to calculate the Eucledian distance between two given points.
The points are represented as co-ordinates on the Cartesian Plane.

Basic Theory:




> On a Cartesian Plane the points are represented as *(x,y)*  
where, *x* is called **abscissa** and *y* is called **ordinate** .  
* *x*-coordinate (abscissa) the distance from a point to the vertical or y -axis, measured parallel to the horizontal or x -axis.
* *y*-coordinate (ordinate) the distance from a point to the vertical or x-axis, measured parallel to the horizontal or x-axis.


 >The formula for calculating the Euclidean Distance is as follow:
> *distance = √[(x₂-x₁)² + (y₂-y₁)²]*
  

In this program we will use numpy *arrays* for the storing and representing the
cartesian points

```python
#importing libraries
import numpy as np
from rich import print
import math
```

```python
#initializing lists to store coordinate values
point1 = []
point2 = []
```

```python
#getting user inputs

#for point1
print("Enter the values of Point 1 : ")
for i in range(2):
  point1.append(float(input()))

#for point2
print("Enter the values of Point 2 : ")
for i in range(2):
  point2.append(float(input()))
```

```python
#converting lists into numpy arrays
point_1 = np.array(point1)
point_2 = np.array(point2)
```

```python
#The values of Point 1:
print(f'The values of point 1 are : {point_1}')
#values of point 2:
print(f'The values of point 2 are : {point_2}')
```

```python
def distance(p1,p2):
  return np.sqrt(np.sum((p1-p2)**2))              # The sqrt() function is used to calculate the square root of the value
```

```python
#calculating the Euclidean distance

dist = distance(point_1,point_2)

print(f'The Distance between points is {math.ceil(dist)} units.')
                                                  # the elements of both array will be subtracted according to there indexes
                                                  # the subracted values will be squared and then added
                                                  # then the square root is calculated
```

**Graphical Representation of the above process**

```python
import matplotlib.pyplot as plt

x1 = point_1[0]
y1 = point_1[1]

x2 = point_2[0]
y2 = point_2[1]

plt.scatter(x1, y1, label='Point 1')
plt.scatter(x2, y2, label='Point 2')

# Plot the line segment between point_1 and point_2
plt.plot([x1, x2], [y1, y2], color='red', linestyle='--', label='Line Segment')

# Add labels to the points
plt.text(x1, y1, f' ({x1},{y1})', fontsize=9, ha='right')
plt.text(x2, y2, f' ({x2},{y2})', fontsize=9, ha='right')

# Add text along the line segment for distance
mid_x = (x1 + x2) / 2
mid_y = (y1 + y2) / 2
plt.text(mid_x, mid_y, f' Distance: {dist:.2f}', color='black', fontsize=10, va='bottom')

plt.title("Scatter Plot on Cartesian Plane")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.axis('equal') # Ensures equal scaling
plt.legend()
plt.show()
```
