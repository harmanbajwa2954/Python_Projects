# Distance Calculator


This application aims to calculate the Eucledian distance between two given points.
The points are represented as co-ordinates on the Cartesian Plane.

Basic Theory:




> On a Cartesian Plane the points are represented as *(x,y)*  
where, *x* is called **abscissa** and *y* is called **ordinate** .  
* *x*-coordinate (abscissa) the distance from a point to the vertical or y -axis, measured parallel to the horizontal or x -axis.
* *y*-coordinate (ordinate) the distance from a point to the vertical or x-axis, measured parallel to the horizontal or x-axis.


 >The formula for calculating the Euclidean Distance is as follow:
> *distance = √[(x₂-x₁)² + (y₂-y₁)²]*
